# 9. Palindrome Number

Given an integer x, return true if x is a palindrome,and false otherwise.

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        string = str(x)
        return string == string[::-1]
```

```python
solver = Solution

num = 121
print(f"Number {num}, is Palindrome: {solver.isPalindrome(solver, num)}")
num = -121
print(f"Number {num}, is Palindrome: {solver.isPalindrome(solver, num)}")
num = 10
print(f"Number {num}, is Palindrome: {solver.isPalindrome(solver, num)}")

num = 1234
print(f"Number {num}, is Palindrome: {solver.isPalindrome(solver, num)}")
```
