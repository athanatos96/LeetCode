# 234. Palindrome Linked List

Given the head of a singly linked list, return true if it is a palindrome or false otherwise.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        def convertListNode2List(node):
            list_val = []
            while node != None:
                value = node.val
                node = node.next
                
                list_val.append(value)
            return list_val
        
        li = convertListNode2List(head)
        
        return li == li[::-1]
```

```python
def convertNum2List(ni):
    le = len(ni)
    node = None
    next_node = None
    for i in range(le-1, -1, -1):
        node = ListNode(val= ni[i], next = next_node)
        next_node = node
    return node
def convertList2Num(node):
    list_val = []
    while node != None:
        value = node.val
        node = node.next
        
        list_val.append(value)
    return list_val
```

```python
solver = Solution

head = convertNum2List([1,2,2,1])
print(f"head: {convertList2Num(head)}, Output: {solver.isPalindrome(solver, head)}")

head = convertNum2List([1,2])
print(f"head: {convertList2Num(head)}, Output: {solver.isPalindrome(solver, head)}")
```
