# 206. Reverse Linked List

Given the head of a singly linked list, reverse the list, and return the reversed list.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        curr = head

        while curr:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next

        return prev
```

```python
def convertNum2List(ni):
    le = len(ni)
    node = None
    next_node = None
    for i in range(le-1, -1, -1):
        node = ListNode(val= ni[i], next = next_node)
        next_node = node
    return node
def convertList2Num(node):
    list_val = []
    while node != None:
        value = node.val
        node = node.next
        
        list_val.append(value)
    return list_val
```

```python
solver = Solution

head = convertNum2List([1,2,3,4,5])
print(f"head: {convertList2Num(head)}, Output: {convertList2Num(solver.reverseList(solver, head))}")

head = convertNum2List([1,2])
print(f"head: {convertList2Num(head)}, Output: {convertList2Num(solver.reverseList(solver, head))}")

head = convertNum2List([])
print(f"head: {convertList2Num(head)}, Output: {convertList2Num(solver.reverseList(solver, head))}")
```
