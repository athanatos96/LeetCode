# 14. Longest Common Prefix

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        minword = min(strs, key=len)
        leng = len(minword)
        i = 0 
        while i < len(strs):
            if strs[i][0:leng] == minword:
                i+=1
            else:
                minword = minword[0:leng-1]
                leng -= 1

            if(minword ==""):
                break
        return minword
```

```python
solver = Solution

strs = ["flower","flow","flight"]

print(f"List : {strs} -- > \"fl\". Output: \"{solver.longestCommonPrefix(solver, strs)}\"")

strs = ["dog","racecar","car"]

print(f"Roman Number: {s} -- > \"\". Output: \"{solver.longestCommonPrefix(solver, strs)}\"")
```
