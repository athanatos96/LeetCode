# 171. Excel Sheet Column Number

Given a string columnTitle that represents the column title as appears in an Excel sheet, return its corresponding column number.

For example:
```
A -> 1
B -> 2
C -> 3
...
Z -> 26
AA -> 27
AB -> 28 
...
```

````python
class Solution:
    def titleToNumber(self, columnTitle: str) -> int:
        exponent = len(columnTitle) - 1 
        count = 0
        letters = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
        dictionary = dict(zip(letters, [i for i in range(1, 27)]))
        for i in columnTitle:
            count += dictionary[i] * 26** exponent
            exponent -= 1
        return count
````

````python
solver = Solution

columnTitle = "A"
print(f"columnTitle: {columnTitle}, Output: {solver.titleToNumber(solver, columnTitle)}")

columnTitle = "AB"
print(f"columnTitle: {columnTitle}, Output: {solver.titleToNumber(solver, columnTitle)}")

columnTitle = "ZY"
print(f"columnTitle: {columnTitle}, Output: {solver.titleToNumber(solver, columnTitle)}")
````
