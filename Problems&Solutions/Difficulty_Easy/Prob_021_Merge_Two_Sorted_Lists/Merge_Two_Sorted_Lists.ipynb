# 21. Merge Two Sorted Lists

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:

        ini = ListNode(val = -150)
        current = ini
        while list1 != None or list2 != None:

            elem1 = list1.val if list1!=None else 150
            elem2 = list2.val if list2!=None else 150

            #print(f"Elem1: {elem1}, Elem2: {elem2}")

            if elem1<elem2:
                current.next = ListNode(val=elem1)
                list1 = list1.next if list1 != None else None
            else:
                current.next = ListNode(val=elem2)
                list2 = list2.next if list2 != None else None
            
            current = current.next

            
        return ini.next
```

```python
def convertNum2List(ni):
    le = len(ni)
    node = None
    next_node = None
    for i in range(le-1, -1, -1):
        node = ListNode(val= ni[i], next = next_node)
        next_node = node
    return node
def convertList2Num(node):
    list_val = []
    while node != None:
        value = node.val
        node = node.next
        
        list_val.append(value)
    return list_val

solver = Solution

list1 = convertNum2List([1,2,4])
list2 = convertNum2List([1,3,4])
print(f"list1: {convertList2Num(list1)}, list2: {convertList2Num(list2)}, Output: {convertList2Num(solver.mergeTwoLists(solver, list1, list2))}")

list1 = convertNum2List([])
list2 = convertNum2List([])
print(f"list1: {convertList2Num(list1)}, list2: {convertList2Num(list2)}, Output: {convertList2Num(solver.mergeTwoLists(solver, list1, list2))}")

list1 = convertNum2List([])
list2 = convertNum2List([0])
print(f"list1: {convertList2Num(list1)}, list2: {convertList2Num(list2)}, Output: {convertList2Num(solver.mergeTwoLists(solver, list1, list2))}")
```
