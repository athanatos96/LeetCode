# 23. Merge k Sorted Lists

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        nodes = []
        head = point = ListNode(0)
        
        for l in lists:
            while l:
                nodes.append(l.val)
                l = l.next

        for x in sorted(nodes):
            point.next = ListNode(x)
            point = point.next

        return head.next
```

```python
# Aux funtions to create ListNode from Lists and to convert ListNodes to Lists
def convertNum2List(ni):
    le = len(ni)
    node = None
    next_node = None
    for i in range(le-1, -1, -1):
        node = ListNode(val= ni[i], next = next_node)
        next_node = node
    return node
def convertList2Num(node):
    list_val = []
    while node != None:
        value = node.val
        node = node.next
        
        list_val.append(value)
    return list_val

def listNum2listListNode(l):
    output = []
    for elem in l:
        output.append(convertNum2List(elem))
    return output
def listListNode2listNum(l):
    output = []
    for elem in l:
        output.append(convertList2Num(elem))
    return output
    
```

```python
solver = Solution

lists = [[1,4,5],[1,3,4],[2,6]]
lists = listNum2listListNode(lists)
print(f"lists: {listListNode2listNum(lists)}, Output: {convertList2Num(solver.mergeKLists(solver, lists))}")


lists = []
lists = listNum2listListNode(lists)
print(f"lists: {listListNode2listNum(lists)}, Output: {convertList2Num(solver.mergeKLists(solver, lists))}")

lists = [[]]
lists = listNum2listListNode(lists)
print(f"lists: {listListNode2listNum(lists)}, Output: {convertList2Num(solver.mergeKLists(solver, lists))}")

lists = [[1,4,5],[1,2,3,4],[2,6,10]]
lists = listNum2listListNode(lists)
print(f"lists: {listListNode2listNum(lists)}, Output: {convertList2Num(solver.mergeKLists(solver, lists))}")
```
