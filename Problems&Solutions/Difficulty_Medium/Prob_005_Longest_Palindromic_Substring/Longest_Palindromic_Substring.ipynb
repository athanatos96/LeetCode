# 5. Longest Palindromic Substring

Given a string s, return the longest 
palindromic substring in s.

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        
        def checkPoli(wor):
            return wor == wor[::-1] 

        polyndroms=[]

        for i in range(len(s)):
            for j  in range(i,len(s)+1):
                if s[i] == s[j-1]:
                    if checkPoli(s[i:j]):
                        polyndroms.append( s[i:j])

        return max(polyndroms, key=len)
```

```python
solver = Solution
palindrome_words = ["madam", "racecar", "level", "civic", "deified", "pop", "wow", "rotator", "refer", "redivider", "repaper", "reviver", "rotor", "sagas", "solos", "stats", "tenet", "redder","madamimadam","deified","deified","civic","level","radar","wow","pop","mom","dad","noon","eye","level","madam","racecar","civic","sagas","solos","stats","tenet","redder","redivider","repaper","reviver","rotor"]
non_palindrome_words = ["hello", "goodbye", "cat", "dog", "house", "car", "book", "computer", "table", "desk", "chair", "pen", "pencil", "paper", "keyboard", "monitor", "mouse", "telephone", "television", "furniture", "appliance", "tree", "flower", "plant", "fruit", "vegetable", "sofa", "couch", "ottoman", "stool", "bed", "dresser", "nightstand", "lamp", "picture", "frame", "mirror", "clock", "watch", "jewelry", "ring", "necklace", "bracelet", "earrings", "watch", "belt", "bag", "purse", "wallet", "shoe", "boot", "sandal", "loafers", "sneakers", "hat", "cap", "scarf", "gloves", "mittens", "coffee", "tea", "juice", "water", "milk", "soda", "beer", "wine", "champagne", "whiskey", "vodka", "gin", "rum", "cognac", "brandy", "liquor", "cocktail", "juice", "food", "cooking", "restaurant", "kitchen", "baking", "dinner", "lunch", "breakfast", "recipe", "ingredient", "spice", "salt", "sugar", "pepper", "flour", "yeast", "egg", "milk", "butter", "cheese", "chocolate", "candy", "cookie", "pie", "cake", "bread"]

print("-------------------------------------")
print("Palindrome Words")
print("-------------------------------------")
for palindrome in palindrome_words:
    print(f"Longest Palindrome in {palindrome} is: {solver.longestPalindrome(solver,palindrome)}")

print("-------------------------------------")
print("Non Palindrome Words")
print("-------------------------------------")
for non_palindrome in non_palindrome_words:
    print(f"Longest Palindrome in {non_palindrome} is: {solver.longestPalindrome(solver,non_palindrome)}")
```
