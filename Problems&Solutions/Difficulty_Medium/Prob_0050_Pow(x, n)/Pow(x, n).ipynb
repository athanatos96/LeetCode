# 50. Pow(x, n)

Implement pow(x, n), which calculates x raised to the power n (i.e., xn).

```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        # Easy aproach
        # return x ** n

        # Hard approach
        if n==0: return 1
        sol = x
        abs_n = abs(n)
        jump = x
        count = 1
        while abs_n > 1:
            jump = x
            count = 1
            while abs_n-count > 0:
                sol *= jump
                abs_n -= count
                jump *= x 
                count += 1
        return sol if n>0 else 1/sol
```

```python
solver = Solution

x = 2.00000
n = 10
print(f"x: {x}, n: {n}, Output: {solver.myPow(solver, x, n)}")

x = 2.10000
n = 3
print(f"x: {x}, n: {n}, Output: {solver.myPow(solver, x, n)}")

x = 2.00000
n = -2
print(f"x: {x}, n: {n}, Output: {solver.myPow(solver, x, n)}")

x = 2.00000
n = 0
print(f"x: {x}, n: {n}, Output: {solver.myPow(solver, x, n)}")

x = 0.00001
n = 2147483647 # 2**31 -1
print(f"x: {x}, n: {n}, Output: {solver.myPow(solver, x, n)}")
```
