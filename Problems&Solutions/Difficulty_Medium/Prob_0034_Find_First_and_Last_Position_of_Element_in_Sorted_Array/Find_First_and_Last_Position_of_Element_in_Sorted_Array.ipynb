# 34. Find First and Last Position of Element in Sorted Array

Given an array of integers nums sorted in non-decreasing order, find the starting and ending position of a given target value.

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if len(nums)==0:
            return [-1,-1]

        def find_range(l, r, side):
            pointer = l + (r-l)//2
            while True:
                if nums[pointer] == target:
                    if side =='Left':
                        if pointer > 0 and nums[pointer - 1] != target:
                            break
                        elif pointer == 0:
                            return 0
                    elif side =='Right':
                        if pointer < len(nums) -1 and nums[pointer + 1] != target:
                            break
                        elif pointer == len(nums) -1:
                            return len(nums) -1
                    else:
                        print("ERROR")
                        return -1
                
                if l==r: return -1

                if ((r-l)//2) == 0: 
                    if nums[r]==target:
                        return r
                    elif nums[l]==target:
                        return l
                    else:
                        return -1

                if nums[pointer] < target or (side =='Right' and nums[pointer] == target):
                    # move to Right
                    l = pointer
                elif nums[pointer] > target or (side =='Left' and nums[pointer] == target):
                    # move to left
                    r = pointer
                pointer = l + (r-l)//2
            return pointer

        
        l = 0
        r = len(nums) -1
        left = find_range(l, r, side = 'Left')

        if left == -1:
            return [-1,-1]

        right = find_range(left, r, side = 'Right')
        return [left,right] 
```

```python
solver = Solution

nums = [5,7,7,8,8,10]
target = 8
print(f"nums: {nums}, target: {target}, Output: {solver.searchRange(solver, nums, target)}")

nums = [5,7,7,8,8,10]
target = 6
print(f"nums: {nums}, target: {target}, Output: {solver.searchRange(solver, nums, target)}")

nums = []
target = 0
print(f"nums: {nums}, target: {target}, Output: {solver.searchRange(solver, nums, target)}")
```
