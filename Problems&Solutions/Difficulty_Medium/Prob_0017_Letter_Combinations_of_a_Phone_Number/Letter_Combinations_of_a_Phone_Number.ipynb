# 17. Letter Combinations of a Phone Number

Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent. Return the answer in any order.

A mapping of digits to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:

        dic = { "0": "", "1":"", "2": "abc", "3": "def", "4":"ghi", "5":"jkl", 
               "6":"mno", "7":"pqrs", "8":"tuv", "9":"wxyz"}
        
        res = []
        
        if len(digits) == 0:
            return res
        
        if len(digits) == 1:
            digit = digits[0]
            if len(dic[digit]) == 3:
                res.append(list([dic[digit][0],dic[digit][1],dic[digit][2]]))
            else:
                res.append(list([dic[digit][0],dic[digit][1],dic[digit][2],dic[digit][3]]))
        
        all_combinations = [''] if digits else []
        
        for number in digits:
            current_combinations = list()
            
            for l in dic[number]:
                for combination in all_combinations:
                    current_combinations.append(combination + l)
    
            all_combinations = current_combinations
            
        return all_combinations
```

```python
solver = Solution

digits = "23"
print(f"digits: {digits}, Output: {solver.letterCombinations(solver, digits)}")

digits = ""
print(f"digits: {digits}, Output: {solver.letterCombinations(solver, digits)}")


digits = "2"
print(f"digits: {digits}, Output: {solver.letterCombinations(solver, digits)}")

```

```python
'''Test with 10 two digit sequences, 10 three digit sequences, 5 four digit sequences'''

import random

random_list = []
for i in range(0, 10):
    random_list.append(random.randint(2, 9))
for i in range(0,10):
    random_list.append(random.randint(10, 99))    
for i in range(0, 10):
    random_list.append(random.randint(100, 999))
for i in range(0, 5):
    random_list.append(random.randint(1000, 9999))

str_nums = list(map(str, random_list))
    
for j in str_nums:
    print(f"Number Combination: {j} Letter Combination: {letterCombinations(j)}")
```
