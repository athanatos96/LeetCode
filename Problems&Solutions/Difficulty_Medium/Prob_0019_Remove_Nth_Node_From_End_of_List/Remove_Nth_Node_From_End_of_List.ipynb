# 19. Remove Nth Node From End of List

Given the head of a linked list, remove the nth node from the end of the list and return its head.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        def convertNum2List(ni):
            le = len(ni)
            node = None
            next_node = None
            for i in range(le-1, -1, -1):
                node = ListNode(val= ni[i], next = next_node)
                next_node = node
            return node
        def convertList2Num(node):
            list_val = []
            while node != None:
                value = node.val
                node = node.next
                
                list_val.append(value)
            return list_val
        
        list_num = convertList2Num(head)
        list_num.pop(len(list_num)-n)

        return convertNum2List(list_num)
```

```python
# Aux funtions to create ListNode from Lists and to convert ListNodes to Lists
def convertNum2List(ni):
    le = len(ni)
    node = None
    next_node = None
    for i in range(le-1, -1, -1):
        node = ListNode(val= ni[i], next = next_node)
        next_node = node
    return node
def convertList2Num(node):
    list_val = []
    while node != None:
        value = node.val
        node = node.next
        
        list_val.append(value)
    return list_val
```

```python
solver = Solution

head = convertNum2List([1,2,3,4,5])
n = 2
print(f"head: {convertList2Num(head)}, Output: {convertList2Num(solver.removeNthFromEnd(solver, head, n))}")

head = convertNum2List([1])
n = 1
print(f"head: {convertList2Num(head)}, Output: {convertList2Num(solver.removeNthFromEnd(solver, head, n))}")

head = convertNum2List([1,2])
n = 1
print(f"head: {convertList2Num(head)}, Output: {convertList2Num(solver.removeNthFromEnd(solver, head, n))}")
```
