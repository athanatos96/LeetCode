## Solution 1

Convert the ListNode into an integer Number perform the addition, and then convert back into a ListNode.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:

    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        def convertList2Num(li, count=0,  total = 0):
            value = li.val
            next_node = li.next

            total2 = total + value * 10**count

            if(next_node==None):
                return total2
            else:
                return convertList2Num(next_node, count = count+1, total = total2)

        def convertNum2List(ni):
            le = len(str(ni))
            prev = None
            for i in range(le):
                node = ListNode(val=int(str(ni)[i]), next=prev)
                prev= node
            return node

        num1 = convertList2Num(l1)
        num2 = convertList2Num(l2)

        return convertNum2List(num1 + num2) 
```

## Solution 2

Convert the ListNode into an integer Number perform the addition, and then convert back into a ListNode.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    def setNext(self, next = None):
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        orgininal = None
        carry = 0
        prev_node = None
        while True:
            val1 = 0
            val2 = 0
            c = 2
            if l1 is not None:
                val1 = l1.val
                c-=1
            if l2 is not None:
                val2 = l2.val  
                c-=1

            if(c==2 and carry==0):
                return orgininal
            
            tot = val1 + val2 + carry
            if tot>=10:
                node = ListNode(val = tot-10)
                carry = 1
            else:
                node = ListNode(val = tot)
                carry = 0

            if orgininal is None:
                orgininal = node

            if prev_node is not None:
                prev_node.setNext(next = node)
            prev_node = node

            if l1 is not None:
                l1 = l1.next 
            if l2 is not None:
                l2 = l2.next 
```

```python
l1 = convertNum2List(342)
l2 = convertNum2List(465)

solver = Solution
convertList2Num(solver.addTwoNumbers(solver,l1,l2))
```
