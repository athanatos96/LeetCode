# 11. Container With Most Water

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        leng = min(len(height), 10**5)
        max_area = 0
        i = 0
        j = leng-1
        size_left = True
        while i<j:
            val1 = height[i]
            val2 = height[j]
            #print(f"Comparison Pos: {i}, {j}")

            new_area = (j-i) * min(val1,val2)
            if new_area > max_area:
                 max_area = new_area

            if size_left:
                if val1 >= val2:
                    j -= 1
                    size_left = False
                    continue
                i += 1
            else:
                if val2 >= val1:
                    i += 1
                    size_left = True
                    continue
                j -= 1
        return max_area
```

```python
solver = Solution


height = [1,8,6,2,5,4,8,3,7]

print(f"List: {height}, Solution: 49, Output: {solver.maxArea(solver, height)}")

height = [1,1]

print(f"List: {height}, Solution: 1, Output: {solver.maxArea(solver, height)}")
```
