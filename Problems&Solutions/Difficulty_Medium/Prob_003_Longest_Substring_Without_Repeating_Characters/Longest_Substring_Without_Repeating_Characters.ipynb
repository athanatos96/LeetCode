# 3. Longest Substring Without Repeating Characters


Given a string s, find the length of the longest substring without repeating characters.

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:

        leng = len(s)
        substring = []
        record = 0

        for i in range(leng):

            for y in range(leng-i):
                j = i + y
                elem = s[j]
                if elem in substring:
                    break
                substring.append(elem)

            nn = len(substring)
            if nn>record:
                record=nn
            substring = []

        return record
```

```python
s1 = "abcabcbb"
s2 = "bbbbb"
s3 = "pwwkew"



solver = Solution
print(f"For String \"{s1}\" \t --> {solver.lengthOfLongestSubstring(solver, s1)}")
print(f"For String \"{s2}\" \t --> {solver.lengthOfLongestSubstring(solver, s2)}")
print(f"For String \"{s3}\" \t --> {solver.lengthOfLongestSubstring(solver, s3)}")
```
